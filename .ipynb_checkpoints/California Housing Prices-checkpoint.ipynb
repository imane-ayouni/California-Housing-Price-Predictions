# California housing price predictions

## Topic
1. longitude: A measure of how far west a house is; a higher value is farther west

2. latitude: A measure of how far north a house is; a higher value is farther north

3. housingMedianAge: Median age of a house within a block; a lower number is a newer building

4. totalRooms: Total number of rooms within a block

5. totalBedrooms: Total number of bedrooms within a block

6. population: Total number of people residing within a block

7. households: Total number of households, a group of people residing within a home unit, for a block

8. medianIncome: Median income for households within a block of houses (measured in tens of thousands of US Dollars)

9. medianHouseValue: Median house value for households within a block (measured in US Dollars)

10. oceanProximity: Location of the house w.r.t ocean/sea

## Feature engineering

```python
housing["rooms_per_household"] = housing["total_rooms"]/housing["households"]
housing["bedrooms_per_room"] = housing["total_bedrooms"]/housing["total_rooms"]
housing["population_per_household"]=housing["population"]/housing["households"]
housing["income_per_household"] = housing["households"]/housing["median_income"]
housing["rooms_per_bedroom"] = housing["total_rooms"]/housing["total_bedrooms"]
corr_matrix = housing.corr()
corr_matrix["median_house_value"].sort_values(ascending=False)
```

The new features show better linear correlations than the some of the original ones:
- rooms per bedroom has a significant positive correlation with the target and bedrooms per room have a strong negtive one, meaning that the more rooms there are in a house, the more expensive it tends to be.
- rooms per household also has a somewhat significant positive correlation.
- population per household has a slight negative correlation, meaning that the households who have fewer members have higher property values.
- weirdly, the income per household shows a negative correlation with the target, which means that the less one family's household, the more valuable their house can be.

#### Spliting training and testing sets

```python
housing["income_category"] = income_category
split = StratifiedShuffleSplit(n_splits= 1 , test_size= 0.2 , random_state=42)
for train_index, test_index in split.split(housing, housing["income_category"]):
    train_set = housing.loc[train_index]
    test_set = housing.loc[test_index]
```

```python
housing_copy["income_category"] = income_category
for train_index, test_index in split.split(housing_copy, housing_copy["income_category"]):
    train_copy = housing_copy.loc[train_index]
    test_copy = housing_copy.loc[test_index]
```

```python
for s in (train_set , test_set, train_copy, test_copy):
    s.drop("income_category" , axis = 1 ,  inplace = True)
```

In here, Stratified split is applied based on the feature of income category, which in fact is an additional feature that cuts the median income feature into 5 categories (which was used earlier). The feature median income is the most correlated one with the target, and seems to have a lot of importance. So, we used stratified split to make sure the train and test sets have a balanced amount of each of the five categories, than we deleted the income category column.

```python
labels_train_copy = train_copy["median_house_value"].copy()
train_copy = train_copy.drop(["median_house_value"], axis = 1)
labels_test_copy = test_copy["median_house_value"].copy()
test_copy = test_copy.drop(["median_house_value"],axis = 1)
```

In this part, we separated the labels from the datasets. We used the modified dataset in which we added additional columns (that had better correlations with the target), and plus we will try with original data set as well to see whether or not the new features have any impact on the model's performance. 

#### Preparing data for ML algorithms

```python
train_set_num = train_set.drop("ocean_proximity", axis = 1)
test_set_num = test_set.drop("ocean_proximity",axis = 1)
```

```python
train_copy_num = train_copy.drop("ocean_proximity", axis = 1)
test_copy_num = test_copy.drop("ocean_proximity", axis = 1) 
```

here we dropped the ocean proximity column because it is a categorical attribute and won't be treated like the numerical ones during data transformations.

```python
num_pipeline = Pipeline([('imputer', SimpleImputer(strategy="median")),('std_scaler', StandardScaler())])
num_attr = list(train_set_num)
cat_attr = ["ocean_proximity"]
full_pipeline = ColumnTransformer([("num", num_pipeline, num_attr),("cat", OneHotEncoder(), cat_attr),])
train_prepared = full_pipeline.fit_transform(train_set)
```

```python
num_attr = list(test_set_num)
cat_attr = ["ocean_proximity"]
full_pipeline = ColumnTransformer([("num", num_pipeline, num_attr),("cat", OneHotEncoder(), cat_attr),])
test_prepared = full_pipeline.fit_transform(test_set)
```

```python
num_attr = list(train_copy_num)
cat_attr = ["ocean_proximity"]
full_pipeline = ColumnTransformer([("num", num_pipeline, num_attr),("cat", OneHotEncoder(), cat_attr),])
train_copy_prepared = full_pipeline.fit_transform(train_copy)
```

```python
num_attr = list(test_copy_num)
cat_attr = ["ocean_proximity"]
full_pipeline = ColumnTransformer([("num", num_pipeline, num_attr),("cat", OneHotEncoder(), cat_attr),])
test_copy_prepared = full_pipeline.fit_transform(test_copy)
```

In this part, we used Sklearn's pipelines and column transformer to transform the totality of the sets at the same time, the transformations applied are the following:
- Imputing empty values by the median for numerical values (specifically for total bedrooms column)
- Scaling the numerical data using the standard scaler which is more efficient than the Min-Max scaler at handling outliers.
- OneHot encoding for categorical attribute because the ocean proximity feature contains more than 2 categories.
